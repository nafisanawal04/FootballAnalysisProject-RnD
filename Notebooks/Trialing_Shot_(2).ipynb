```python
import cv2
import json
import numpy as np
from google.colab.patches import cv2_imshow

# Load the predictions from the JSON file
with open('/content/SoccerNetv2-DevKit/Task1-ActionSpotting/CALF/inference/outputs/Predictions-v2.json', 'r') as file:
    predictions = json.load(file)['predictions']

# Define a helper function to convert game time to seconds
def game_time_to_seconds(game_time):
    half, time = game_time.split(" - ")
    minutes, seconds = map(int, time.split(":"))
    return (int(half) - 1) * 45 * 60 + minutes * 60 + seconds  # Each half is 45 minutes

# Open the video file
video_path = '/content/video1.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print(f"📹 Video Resolution: {frame_width}x{frame_height}")

# Set a fixed large font size for high resolution videos
font_scale = 2.0  # Larger font for better visibility
font_thickness = 4  # Thicker text for readability

# Duration to show each annotation (in seconds)
annotation_duration = 2  # 2 seconds
annotation_duration_frames = int(annotation_duration * fps)

# Map predictions to frame ranges
frame_predictions = {}
for prediction in predictions:
    timestamp = game_time_to_seconds(prediction['gameTime'])
    frame_number = int(timestamp * fps)
    # Make annotation visible for the duration
    for i in range(annotation_duration_frames):
        frame_predictions[frame_number + i] = prediction

# Prepare output video
output_path = "/content/output_video5.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Process every frame in the video
frame_idx = 0
print("Processing video...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Check if there's a prediction for this frame
    if frame_idx in frame_predictions:
        prediction = frame_predictions[frame_idx]
        label = prediction['label']

        # Only show the label (no game time)
        text = label

        # Get text size for the background rectangle
        font = cv2.FONT_HERSHEY_SIMPLEX
        text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
        text_width, text_height = text_size

        # Set rectangle position and dynamic padding
        padding = 20
        rect_width = text_width + (padding * 2)
        rect_height = text_height + (padding * 2)

        # Set position from top
        top_margin = 30
        rect_x = (frame_width - rect_width) // 2
        rect_y = top_margin

        # Draw a background rectangle for the text (white)
        cv2.rectangle(frame,
                     (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     (255, 255, 255),  # White background
                     -1)  # Filled rectangle

        # Draw a border around the rectangle (black)
        cv2.rectangle(frame,
                     (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     (0, 0, 0),  # Black border
                     2)  # Border thickness

        # Add the label text on top of the rectangle
        text_x = rect_x + padding
        text_y = rect_y + padding + text_height
        cv2.putText(frame, text, (text_x, text_y),
                   font, font_scale, (0, 0, 0), font_thickness, cv2.LINE_AA)

    # Write every frame to output
    out.write(frame)
    frame_idx += 1

    # Print progress every 100 frames
    if frame_idx % 100 == 0:
        progress = (frame_idx / total_frames) * 100
        print(f"Processed {frame_idx}/{total_frames} frames ({progress:.1f}%)")

cap.release()
out.release()

print(f"✅ Output video saved at {output_path}")
print(f"📊 Total frames processed: {frame_idx}")
print(f"📝 Total annotations: {len(predictions)}")
```

#Code for free kick efficiency percentage

```python
import cv2
import json
import numpy as np
from collections import defaultdict

# Load the predictions from the JSON file
with open('/content/SoccerNetv2-DevKit/Task1-ActionSpotting/CALF/inference/outputs/Predictions-v2.json', 'r') as file:
    predictions = json.load(file)['predictions']

# Define a helper function to convert game time to seconds
def game_time_to_seconds(game_time):
    half, time = game_time.split(" - ")
    minutes, seconds = map(int, time.split(":"))
    return (int(half) - 1) * 45 * 60 + minutes * 60 + seconds

# ========== FREE KICK EFFICIENCY CALCULATION ==========
def calculate_free_kick_efficiency(predictions):
    """
    Calculate free kick efficiency based on rule-based algorithm
    """
    # Sort predictions by timestamp
    sorted_predictions = sorted(predictions, key=lambda x: game_time_to_seconds(x['gameTime']))

    free_kick_data = []
    total_free_kicks = 0
    total_points = 0

    # INCREASED TIME WINDOW - Changed from 15 to 30 seconds
    TIME_WINDOW = 30  # 30 seconds after free kick to capture more outcomes

    # Identify all free kicks and analyze outcomes
    for i, pred in enumerate(sorted_predictions):
        label = pred['label'].lower()

        # Check if this is a free kick event (both direct and indirect)
        if ('free' in label and 'kick' in label) or 'free-kick' in label:
            total_free_kicks += 1
            fk_time = game_time_to_seconds(pred['gameTime'])
            points = 0
            outcome_labels = []

            print(f"\n🔍 Analyzing Free Kick at {pred['gameTime']} (Team: {pred['team']})")

            # Look ahead for outcomes within time window
            for j in range(i + 1, len(sorted_predictions)):
                next_pred = sorted_predictions[j]
                next_time = game_time_to_seconds(next_pred['gameTime'])
                next_label = next_pred['label'].lower()
                time_diff = next_time - fk_time

                # Debug: Show what events we're checking
                print(f"  ⏱️  Checking event at +{time_diff}s: {next_pred['label']}")

                # Stop if we're outside the time window
                if time_diff > TIME_WINDOW:
                    print(f"  ⛔ Outside time window ({TIME_WINDOW}s)")
                    break

                # Check for goal (3 points)
                if 'goal' in next_label:
                    points = max(points, 3)
                    outcome_labels.append(f"Goal (+3pts)")
                    print(f"  ✅ GOAL FOUND! (+3pts)")
                    break  # Goal is the best outcome, stop looking

                # Check for shot on target (1 point) - Fixed label matching
                elif 'shot' in next_label and ('target' in next_label or 'on target' in next_label):
                    if points < 1:
                        points = 1
                        outcome_labels.append(f"Shot on target (+1pt)")
                        print(f"  ✅ Shot on target found! (+1pt)")

                # Check for key pass or corner (0.5 points)
                elif 'pass' in next_label or 'corner' in next_label or 'clearance' in next_label:
                    if points < 0.5:
                        points = 0.5
                        outcome_labels.append(f"Key play (+0.5pt)")
                        print(f"  ✅ Key play found! (+0.5pt)")

                # Check for ball out of play (0 points but still note it)
                elif 'ball out of play' in next_label:
                    if not outcome_labels:  # Only add if no other outcome found
                        outcome_labels.append(f"Ball out of play (0pts)")
                        print(f"  ⚪ Ball out of play")

            total_points += points

            # Store free kick data for annotation
            free_kick_data.append({
                'timestamp': fk_time,
                'gameTime': pred['gameTime'],
                'points': points,
                'outcome': outcome_labels if outcome_labels else ['No significant outcome'],
                'label': pred['label']
            })

            print(f"  📊 Final points for this FK: {points}")

    # Calculate efficiency
    efficiency = (total_points / total_free_kicks * 100) if total_free_kicks > 0 else 0

    return {
        'free_kicks': free_kick_data,
        'total_free_kicks': total_free_kicks,
        'total_points': total_points,
        'efficiency': efficiency
    }

# Calculate free kick efficiency
fk_efficiency = calculate_free_kick_efficiency(predictions)

print("\n" + "="*60)
print("FREE KICK EFFICIENCY ANALYSIS")
print("="*60)
print(f"Total Free Kicks: {fk_efficiency['total_free_kicks']}")
print(f"Total Points: {fk_efficiency['total_points']}")
print(f"Efficiency Score: {fk_efficiency['efficiency']:.2f}%")
print("="*60 + "\n")

for i, fk in enumerate(fk_efficiency['free_kicks'], 1):
    print(f"Free Kick #{i} at {fk['gameTime']}")
    print(f"  Label: {fk['label']}")
    print(f"  Points: {fk['points']}")
    print(f"  Outcome: {', '.join(fk['outcome'])}")
    print()

# ========== VIDEO PROCESSING ==========

# Open the video file
video_path = '/content/BAR_RMA_video.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print(f"📹 Video Resolution: {frame_width}x{frame_height}")
print(f"📹 Video FPS: {fps}")

# Set font properties
font_scale = 2.0
font_thickness = 4
small_font_scale = 1.0
small_font_thickness = 2

# Duration to show each annotation (in seconds)
annotation_duration = 3  # 3 seconds for better visibility
annotation_duration_frames = int(annotation_duration * fps)

# Map predictions to frame ranges
frame_predictions = {}
for prediction in predictions:
    timestamp = game_time_to_seconds(prediction['gameTime'])
    frame_number = int(timestamp * fps)
    for i in range(annotation_duration_frames):
        frame_predictions[frame_number + i] = prediction

# Map free kick efficiency data to frames
frame_freekick_data = {}
for fk in fk_efficiency['free_kicks']:
    frame_number = int(fk['timestamp'] * fps)
    for i in range(annotation_duration_frames):
        frame_freekick_data[frame_number + i] = fk

# Map outcome events to frames (to show points when outcome happens)
frame_outcome_data = {}
frame_efficiency_update = {}  # Track efficiency changes over time

# Track cumulative statistics
cumulative_fks = 0
cumulative_points = 0

for fk in fk_efficiency['free_kicks']:
    fk_time = fk['timestamp']
    fk_frame = int(fk_time * fps)

    # At the free kick moment, increment FK count but no points yet
    cumulative_fks += 1
    for i in range(annotation_duration_frames):
        frame_efficiency_update[fk_frame + i] = {
            'total_fks': cumulative_fks,
            'total_points': cumulative_points,
            'efficiency': 0.0  # 0% until outcome happens
        }

    # Find the outcome event in predictions
    for pred in predictions:
        pred_time = game_time_to_seconds(pred['gameTime'])
        time_diff = pred_time - fk_time
        pred_label = pred['label'].lower()

        # Check if this prediction is the outcome within 30 seconds
        if 0 < time_diff <= 30:
            is_outcome = False
            outcome_text = ""
            outcome_points = 0

            if 'goal' in pred_label:
                is_outcome = True
                outcome_text = "GOAL from Free Kick!"
                outcome_points = 3
            elif 'shot' in pred_label and 'target' in pred_label:
                is_outcome = True
                outcome_text = "Shot on Target from FK!"
                outcome_points = 1
            elif 'pass' in pred_label or 'corner' in pred_label or 'clearance' in pred_label:
                is_outcome = True
                outcome_text = "Key Play from FK!"
                outcome_points = 0.5
            elif 'Ball out of play' in pred_label:
                is_outcome = True
                outcome_text = "Ball out of play from FK"
                outcome_points = 0.5

            if is_outcome:
                # Update cumulative points when outcome happens
                cumulative_points += outcome_points
                new_efficiency = (cumulative_points / cumulative_fks * 100) if cumulative_fks > 0 else 0

                frame_number = int(pred_time * fps)
                for i in range(annotation_duration_frames):
                    frame_outcome_data[frame_number + i] = {
                        'points': outcome_points,
                        'outcome_text': outcome_text,
                        'fk_gametime': fk['gameTime']
                    }
                    # Update efficiency at outcome moment
                    frame_efficiency_update[frame_number + i] = {
                        'total_fks': cumulative_fks,
                        'total_points': cumulative_points,
                        'efficiency': new_efficiency
                    }
                break  # Only map the first (most significant) outcome

# Prepare output video
output_path = "/content/output_video_freekick_efficiency2.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Process every frame in the video
frame_idx = 0
print("\nProcessing video...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # ========== SCORING LEGEND BOX (Top-Right Corner) ==========
    font = cv2.FONT_HERSHEY_SIMPLEX
    legend_font_scale = 0.7
    legend_thickness = 2
    margin = 20
    line_spacing = 35

    # Get current efficiency for this frame (dynamic)
    current_efficiency = 0.0
    if frame_idx in frame_efficiency_update:
        current_efficiency = frame_efficiency_update[frame_idx]['efficiency']
    else:
        # Find the most recent efficiency update before this frame
        for past_frame in sorted([f for f in frame_efficiency_update.keys() if f < frame_idx], reverse=True):
            current_efficiency = frame_efficiency_update[past_frame]['efficiency']
            break

    # Legend content with dynamic efficiency
    legend_lines = [
        ("FREE KICK SCORING", (255, 255, 255)),  # White title
        ("Goal: +3 pts", (0, 255, 0)),            # Green
        ("Shot on target: +1 pt", (0, 255, 255)), # Yellow
        ("Key play: +0.5 pt", (0, 165, 255)),     # Orange
        ("No outcome: 0 pt", (0, 0, 200)),        # Red
        ("", (0, 0, 0)),  # Spacer
        (f"FK Efficiency: {current_efficiency:.1f}%", (255, 255, 255))  # Dynamic efficiency
    ]

    # Calculate box dimensions
    max_text_width = 0
    for text, _ in legend_lines:
        if text:  # Skip empty spacer
            text_size = cv2.getTextSize(text, font, legend_font_scale, legend_thickness)[0]
            max_text_width = max(max_text_width, text_size[0])

    box_padding = 15
    box_width = max_text_width + (box_padding * 2)
    box_height = len(legend_lines) * line_spacing + box_padding

    # Position in top-right corner
    box_x = frame_width - box_width - margin
    box_y = margin

    # Draw semi-transparent dark background
    overlay = frame.copy()
    cv2.rectangle(overlay,
                 (box_x, box_y),
                 (box_x + box_width, box_y + box_height),
                 (30, 30, 30),  # Dark background
                 -1)
    # Blend with original frame for transparency
    cv2.addWeighted(overlay, 0.8, frame, 0.2, 0, frame)

    # Draw border
    cv2.rectangle(frame,
                 (box_x, box_y),
                 (box_x + box_width, box_y + box_height),
                 (255, 255, 255),  # White border
                 2)

    # Draw legend text
    current_y = box_y + box_padding + 20
    for text, color in legend_lines:
        if text:  # Skip empty spacer
            cv2.putText(frame, text,
                       (box_x + box_padding, current_y),
                       font, legend_font_scale, color, legend_thickness, cv2.LINE_AA)
        current_y += line_spacing

    # Check if there's a regular prediction for this frame
    if frame_idx in frame_predictions:
        prediction = frame_predictions[frame_idx]
        label = prediction['label']
        text = label

        # Get text size
        text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
        text_width, text_height = text_size

        # Set rectangle position
        padding = 20
        rect_width = text_width + (padding * 2)
        rect_height = text_height + (padding * 2)
        top_margin = 30
        rect_x = (frame_width - rect_width) // 2
        rect_y = top_margin

        # Draw background and border
        cv2.rectangle(frame, (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     (255, 255, 255), -1)
        cv2.rectangle(frame, (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     (0, 0, 0), 2)

        # Add text
        text_x = rect_x + padding
        text_y = rect_y + padding + text_height
        cv2.putText(frame, text, (text_x, text_y),
                   font, font_scale, (0, 0, 0), font_thickness, cv2.LINE_AA)


    # Check if there's outcome data for this frame (show points when outcome happens)
    if frame_idx in frame_outcome_data:
        outcome_data = frame_outcome_data[frame_idx]

        # Create multi-line annotation for outcome
        outcome_label = outcome_data['outcome_text']
        outcome_points = f"Points Earned: +{outcome_data['points']}"

        # Position below other labels
        start_y = 250 if (frame_idx in frame_predictions or frame_idx in frame_freekick_data) else 30
        line_height = 60

        # Color based on points
        if outcome_data['points'] >= 3:
            bg_color = (0, 255, 0)  # Green for goal
        elif outcome_data['points'] >= 1:
            bg_color = (0, 255, 255)  # Yellow for shot
        else:
            bg_color = (0, 165, 255)  # Orange for other

        lines = [outcome_label, outcome_points]

        for idx, line in enumerate(lines):
            # Use larger font for outcome
            display_font_scale = 1.2 if idx == 0 else 0.9
            display_font_thickness = 3

            text_size = cv2.getTextSize(line, font, display_font_scale, display_font_thickness)[0]
            padding = 20
            rect_width = text_size[0] + (padding * 2)
            rect_height = text_size[1] + (padding * 2)
            rect_x = (frame_width - rect_width) // 2
            rect_y = start_y + (idx * line_height)

            # Draw background
            cv2.rectangle(frame, (rect_x, rect_y),
                         (rect_x + rect_width, rect_y + rect_height),
                         bg_color, -1)

            # Draw border
            cv2.rectangle(frame, (rect_x, rect_y),
                         (rect_x + rect_width, rect_y + rect_height),
                         (0, 0, 0), 3)

            # Add text
            text_x = rect_x + padding
            text_y = rect_y + padding + text_size[1]
            cv2.putText(frame, line, (text_x, text_y),
                       font, display_font_scale, (0, 0, 0), display_font_thickness, cv2.LINE_AA)

    # Write frame to output
    out.write(frame)
    frame_idx += 1

    # Print progress
    if frame_idx % 100 == 0:
        progress = (frame_idx / total_frames) * 100
        print(f"Processed {frame_idx}/{total_frames} frames ({progress:.1f}%)")

cap.release()
out.release()

print(f"\n✅ Output video saved at {output_path}")
print(f"📊 Total frames processed: {frame_idx}")
print(f"📝 Total annotations: {len(predictions)}")
print(f"⚽ Free kicks analyzed: {fk_efficiency['total_free_kicks']}")
print(f"🎯 Free kick efficiency: {fk_efficiency['efficiency']:.2f}%")
```

# Goalkeeper Saves Feature
Identify 'shot on target' events not followed by a 'goal' within a 3-second window as 'Goalkeeper Save!' events.

```python
import cv2
import json
import numpy as np
from collections import defaultdict

# Load the predictions from the JSON file
with open('/content/SoccerNetv2-DevKit/Task1-ActionSpotting/CALF/inference/outputs/Predictions-v2.json', 'r') as file:
    predictions = json.load(file)['predictions']

# Define a helper function to convert game time to seconds
def game_time_to_seconds(game_time):
    half, time = game_time.split(" - ")
    minutes, seconds = map(int, time.split(":"))
    return (int(half) - 1) * 45 * 60 + minutes * 60 + seconds

# ========== GOALKEEPER SAVE DETECTION ==========
def detect_goalkeeper_saves(predictions, time_window=3):
    saves = []
    # Sort predictions by timestamp to process events in chronological order
    sorted_predictions = sorted(predictions, key=lambda x: game_time_to_seconds(x['gameTime']))

    for i, pred in enumerate(sorted_predictions):
        label = pred['label'].lower()
        pred_time = game_time_to_seconds(pred['gameTime'])

        # Identify 'shot on target' events
        if 'shots on target' in label:
            is_goal_after_shot = False
            # Look for a 'goal' event within the specified time window
            for j in range(i + 1, len(sorted_predictions)):
                next_pred = sorted_predictions[j]
                next_label = next_pred['label'].lower()
                next_time = game_time_to_seconds(next_pred['gameTime'])
                time_diff = next_time - pred_time

                if 0 < time_diff <= time_window:
                    if 'goal' in next_label:
                        is_goal_after_shot = True
                        break
                elif time_diff > time_window: # Event is too far in the future
                    break

            # If no 'goal' detected after 'shot on target' within the time window, it's a save
            if not is_goal_after_shot:
                saves.append({
                    'label': 'Goalkeeper Save!',
                    'gameTime': pred['gameTime'], # Keep original gameTime for reference
                    'timestamp': pred_time + 2 # Start annotation 2 seconds later
                })
    return saves

# Detect goalkeeper saves
goalkeeper_saves = detect_goalkeeper_saves(predictions)

print(f"Detected {len(goalkeeper_saves)} Goalkeeper Save! events.")
for save in goalkeeper_saves:
    print(f"  - {save['gameTime']} (Annotated at +2s): {save['label']}")

# Combine all predictions for annotation
all_annotations = predictions + goalkeeper_saves

# Sort all annotations by timestamp for consistent processing
all_annotations_sorted = sorted(all_annotations, key=lambda x: x['timestamp'] if 'timestamp' in x else game_time_to_seconds(x['gameTime']))

# ========== VIDEO ANNOTATION ==========

# Open the video file
video_path = '/content/BAR_RMA_save.mp4'  # Ensure this path is correct
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print(f"📹 Video Resolution: {frame_width}x{frame_height}")
print(f"📹 Video FPS: {fps}")

# Set font properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale_general = 1.0
font_thickness_general = 2
font_scale_save = 1.5
font_thickness_save = 3

# Duration to show each annotation (in seconds)
annotation_duration = 2
annotation_duration_frames = int(annotation_duration * fps)

# Map annotations to frame ranges
frame_annotations = defaultdict(list)
for ann in all_annotations_sorted:
    # Use the 'timestamp' field if available (e.g., for Goalkeeper Save! events),
    # otherwise calculate from 'gameTime'.
    event_timestamp = ann['timestamp'] if 'timestamp' in ann else game_time_to_seconds(ann['gameTime'])

    frame_number = int(event_timestamp * fps)
    for i in range(annotation_duration_frames):
        # Store a list of annotations for each frame, as multiple could overlap
        frame_annotations[frame_number + i].append(ann)

# Prepare output video
output_path = "/content/BAR_RMA_save_output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Codec for .mp4 output
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Process every frame in the video
frame_idx = 0
print("\nProcessing video frames for annotation...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    current_frame_annotations = frame_annotations[frame_idx]
    y_offset = 50 # Starting y-position for annotations

    # Display current annotations on the frame
    for ann in current_frame_annotations:
        text = ann['label']

        if text == 'Goalkeeper Save!':
            text_color = (0, 0, 0)  # Black text
            bg_color = (0, 255, 255) # Yellow background
            border_color = (0, 0, 0) # Black border
            current_font_scale = font_scale_save
            current_font_thickness = font_thickness_save
        else:
            # For other labels, use white background with black text and border
            text_color = (0, 0, 0) # Black text
            bg_color = (255, 255, 255) # White background
            border_color = (0, 0, 0) # Black border
            current_font_scale = font_scale_general
            current_font_thickness = font_thickness_general

        # Get text size for the background rectangle
        text_size = cv2.getTextSize(text, font, current_font_scale, current_font_thickness)[0]
        text_width, text_height = text_size

        padding = 15 # Padding around text
        rect_width = text_width + (padding * 2)
        rect_height = text_height + (padding * 2)

        # Center the rectangle horizontally
        rect_x = (frame_width - rect_width) // 2
        rect_y = y_offset

        # Draw a background rectangle
        cv2.rectangle(frame,
                     (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     bg_color, # Background color
                     -1) # Filled rectangle

        # Draw a border around the rectangle
        cv2.rectangle(frame,
                     (rect_x, rect_y),
                     (rect_x + rect_width, rect_y + rect_height),
                     border_color, # Border color
                     2) # Border thickness

        # Add the label text on top of the rectangle
        text_x = rect_x + padding
        text_y = rect_y + padding + text_height
        cv2.putText(frame, text, (text_x, text_y),
                   font, current_font_scale, text_color, current_font_thickness, cv2.LINE_AA)

        # Adjust y_offset for the next annotation to avoid overlap
        y_offset += rect_height + 10 # 10 pixels spacing between annotations

    # Write every frame to output
    out.write(frame)
    frame_idx += 1

    # Print progress every 25 frames
    if frame_idx % 100 == 0:
        progress = (frame_idx / total_frames) * 100
        print(f"Processed {frame_idx}/{total_frames} frames ({progress:.1f}%)")

cap.release()
out.release()

print(f"\n✅ Output video saved at {output_path}")
print(f"📊 Total frames processed: {frame_idx}")
print(f"📝 Total original annotations: {len(predictions)}")
print(f"🧤 Total 'Goalkeeper Save!' annotations: {len(goalkeeper_saves)}")
```
